```python
def calculate_average(scores):
    avg = sum(scores) / len(scores)

    if avg >= 90:
        return "A"
    elif avg >= 80:
        return "B"
    elif avg >= 70:
        return "C"
    elif avg >= 60:
        return "D"
    else:
        return "F"
```

```python
scores = [85, 90, 78, 92]
grade = calculate_average(scores)
print(grade)
```

```python
name = "김철수"
scores = [85, 90, 88]

grade = calculate_average(scores)
print(f"{name}의 등급은 {grade}입니다.")
```

```python
import pandas as pd

books = {
    "도서명": ["데이터분석", "알고리즘", "자바입문", "웹개발", "파이썬"],
    "재고": [12, 20, 8, 3, 15],
    "가격": [35000, 32000, 30000, 28000, 25000]
}

df = pd.DataFrame(books)
print(df)
```

```python
print(df.isnull())
```

```python
df_sorted = df.sort_values(by="가격", ascending=False)
print(df_sorted)
```

```python
df_sorted["할인가격"] = (df_sorted["가격"] * 0.9).astype(int)
print(df_sorted)
```

```python
df_sorted.to_csv("book_data.csv", index=False)
```
